# Create timeseries for electric vehicles to use as flexibility sources in OPF
---


To use electric vehicles in the OPF, we need to create a time series for the electric vehicle consumption. This is done using the `Consumption` class from the `emobpy` package. The `Consumption` class allows us to define the vehicle model, including its specifications and driving behavior.

```python
from emobpy import Mobility
```

Choose the number of vehicles oyou would like to have in your simulation.


```python
n_vehicles =  5
```

Create timeseries profiles for each of the vehicles


```python
mobility_profiles = []

for i in range(n_vehicles):
    m = Mobility(config_folder='config_files')  # Load the mobility profile from the database
    m.set_params(
            name_prefix="BEV1",
            total_hours=168, # one week
            time_step_in_hrs=0.25, # 15 minutes
            category="user_defined",
            reference_date="01/01/2020"
            )
    m.set_stats(
            stat_ntrip_path="TripsPerDay.csv",
            stat_dest_path="DepartureDestinationTrip.csv",
            stat_km_duration_path="DistanceDurationTrip.csv",
            )
    m.set_rules(rule_key="user_defined") # see /config_files/rules.yml, it contains a dictionary
                                    # whose key must be the same as rule_key.
                                    # To see all possible rules `from emobpy.constants import RU
    m.run()
    df = m.timeseries
    mobility_profiles.append(df)
```
